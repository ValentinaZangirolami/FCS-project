L'elemento '=' rappresenta pari merito, per cui si estrapola il numero successivo al carattere.

Mentre per il range si estrapolano i numeri precendenti e successivi al carattere '-' e si considera come valore rappresentativo la media tra il minimo e il massimo.

```python
def transform_string_to_float(x):
    if '-' in str(x):
        patt = re.compile("(\d+)\-(\d+)")
        occ = patt.search(x)
        if occ:
            mean_value=(int(occ[1])+int(occ[2]))/2
            return int(mean_value)
    if '=' in str(x):
        numeric_value=x.replace('=','')
        return int(numeric_value)
```

```python
most_recent['world_rank'] = most_recent['world_rank'].apply(transform_string_to_float)
shanghaidata_recent['world_rank'] = shanghaidata_recent['world_rank'].apply(transform_string_to_float)
```

**Trattazione della variabile university_name**

L'obiettivo è quello di uniformare tutte le variabili *university_name* dei relativi tre dataset, prendendo di riferimento il format costituito dal set di dati *schooldata*. Per semplicità, si definisce una funzione che permette di mantenere quelle università definite correttamente e correggere quelle che differiscono dal nome base.

```python
def nomi_universita(input_uni):
    patt= re.compile(input_uni)
    for valori in schooldata['school_name']:
        if valori==input_uni:
            return input_uni
        else:
            occ=patt.search(valori)
            if occ:
                return valori
            else:
                patt_inverso=re.compile(valori)
                occ_inverso=patt_inverso.search(input_uni)
                if occ_inverso:
                    return valori
```

Si applica la funzione ai tre dataset: *most_recent*, *shanghaidata_recent* e *cwurdata_recent*.

La variabile *num_students* è di tipo *object* a causa delle virgole utilizzate come separatore di migliaia, per cui si eliminano e si convertono in formato *float*.

```python
time_fm['num_students'] = time_fm['num_students'].str.replace(',','').astype(float)
```

Si definiscono due funzioni: *numberfem* e *numbermale* in cui attraverso re.compile si definiscono le posizioni delle percentuali di femmine e maschi nella variabile *female_male_ratio*. Successivamente si trasformano queste quantità in decimali dividendo per 100 e si moltiplicano per il totale di studenti al fine di trovare il numero totale di femmine e il numero totale di maschi.

```python
def numberfem(row):
    
    patt=re.compile('^([0-9]+)')
    occ = patt.search(row['female_male_ratio'])
    
    if occ:
        perc_fm = int(occ.group(1)) / 100
        total_students = row['num_students']
        number_f= round(perc_fm * total_students,0)
        
        return number_f
    
def numbermale(row):
    
    patt=re.compile('([0-9]+)$')
    occ = patt.search(row['female_male_ratio'])
    
    if occ:
        perc_fm = int(occ.group(1)) / 100
        total_students = row['num_students']
        number_m = round(perc_fm * total_students,0)
        
        return number_m
```

I valori trovati vengono aggiunte al dataframe *time_fm*.

```python
time_fm['Female']=time_fm.apply(numberfem,axis=1)
time_fm['Male']=time_fm.apply(numbermale,axis=1)
```

Attraverso la funzione *groupby* è possibile stratificare la somma in base alla variabile *country*. Si creano i due elementi *fem_c* e *male_c*, che contengono rispettivamente il numero di femmine e maschi per ogni stato in base ai dati riferiti all'anno più recente.

```python
fem_c=time_fm.groupby('country')['Female'].sum().copy()
male_c=time_fm.groupby('country')['Male'].sum().copy()
```

Si costituisce una tabella in cui è possibile visualizzare congiuntamente il numero di maschi e di femmine.

```python
dat_5=pd.merge(fem_c, male_c, on='country')
```
